```python
import pandas as pd

# Load Dataset
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df


# Dataset Information
def dataset_info(df):
    print("Shape:", df.shape)
    print("\nColumns:")
    print(df.columns)


# Missing Values
def missing_values(df):
    print("\nMissing Values:")
    print(df.isnull().sum())


# Summary Statistics
def statistics(df):
    print("\nStatistics:")
    print(df.describe())


# Search Records
def search_record(df, column, value):
    return df[df[column] == value]


# Filter by Age
def filter_age(df, age):
    return df[df["age"] > age]


# Filter by Gender
def filter_gender(df, gender):
    return df[df["sex"] == gender]


# Filter by Fare
def filter_fare(df, fare):
    return df[df["fare"] > fare]


# Multiple Filters
def multiple_filter(df, age, gender):
    return df[(df["age"] > age) & (df["sex"] == gender)]


# Sort Data
def sort_data(df, column, ascending=True):
    return df.sort_values(by=column, ascending=ascending)


# Export Data
def export_data(df, file_name):
    df.to_csv(file_name, index=False)
    print(f"Data exported successfully to {file_name}")
```

```python
df = load_data("titanic.csv")

dataset_info(df)

missing_values(df)

statistics(df)

print(filter_age(df, 30))

print(filter_gender(df, "female"))

print(filter_fare(df, 50))

print(multiple_filter(df, 30, "female"))

print(sort_data(df, "fare", False))

filtered_df = filter_age(df, 30)

export_data(filtered_df, "age_above_30.csv")
```
